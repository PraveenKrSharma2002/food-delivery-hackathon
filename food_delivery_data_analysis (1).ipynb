```python
import pandas as pd

df = pd.read_csv("/content/final_food_delivery_dataset.csv")
df.head()
```

```python
def rating_bucket(r):
    if 3.0 <= r <= 3.5:
        return "3.0  3.5"
    elif 3.6 <= r <= 4.0:
        return "3.6  4.0"
    elif 4.1 <= r <= 4.5:
        return "4.1  4.5"
    elif 4.6 <= r <= 5.0:
        return "4.6  5.0"

df["rating_range"] = df["rating"].apply(rating_bucket)
df[["rating", "rating_range"]].head()
```

```python
rating_revenue = (
    df.groupby("rating_range")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

rating_revenue
```

```python
import pandas as pd

df = pd.read_csv("/content/final_food_delivery_dataset.csv")
df.head()
```

```python
restaurant_stats = (
    df.groupby("restaurant_name_y")
      .agg(
          avg_order_value=("total_amount", "mean"),
          total_orders=("order_id", "count")
      )
)

restaurant_stats.head()
```

```python
less_20_orders = restaurant_stats[restaurant_stats["total_orders"] < 20]
less_20_orders.head()
```

```python
less_20_orders.sort_values(
    by="avg_order_value",
    ascending=False
).head(1)
```
